## Grafica de Barras Horizontales
- [Abrir Index](index.ipynb)


```python
plt.figure(figsize = (12, 6)) # Ancho, largo del grafico
marcas_clase = [0, 1, 2, 3, 4, 5]#Datos del eje X
frecuencia =[5, 15, 20, 20, 10, 30] #Datos del eje Y
marcas_textos = ["Coca", "Toronja", "Sprite", "Manzanita", "Punch", "Jarrito"]


# Diagrama de barras con:
# - Barras al 80% (height)
# - Contorno negro ( edgecolor = "k")
plt.barh(marcas_clase, frecuencia,
        height = 0.8, edgecolor = "k", 
        color = ["#740505", "#740560", "#B227C6", "#9363D3", "#6372D3", "#63B1D3" ])
plt.yticks(marcas_clase, marcas_textos, fontsize= 15, rotation= 45)

plt.title("Diagrama de barras", fontsize = 40) # Titulo
plt.xlabel(" absoluta", fontsize = 25) # Etiqueta eje X
plt.ylabel("Marcas de clase", fontsize = 25) # Etiqueta eje Y
plt.grid(True)#Cuatricula
plt.show()#Se muestra grfica en pantalla
```
