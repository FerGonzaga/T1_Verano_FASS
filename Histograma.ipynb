```python
import matplotlib.pyplot as plt
print("Setup terminado")
```

```python
plt.figure(figsize = (12, 6)) ## (ancho, alto) de la gráfica 
 
marcas_clase = [0, 1, 2, 3, 4, 5] # Datos del eje x
frecuencia = [5, 15, 20, 20, 10, 30] # Datos del eje y

plt.bar(marcas_clase, frecuencia, # X y Y en la gráfica
        width = 0.8, # Ancho de las barras
        edgecolor = "k", # Color del borde de barra
        color = ["#FF0000", "#FFF700", "#8FFF00", "#00FFF7", "#000CFF", "#FF00E0"]) # Colores de las barras
plt.xlim(0.0, 6) # Intervalo escala en x
plt.ylim(0, 40) # Intervalo escala en y
plt.grid(True) # Cuadricular gráfica
plt.title("Histograma", fontsize=20) # Título de la gráfica
plt.xlabel('Marcas de clase', fontsize=15) # Etiqueta en x
plt.ylabel('Frecuencia absoluta', fontsize=15) # Etiqueta en y
plt.show() # Imprime la gráfica
```
